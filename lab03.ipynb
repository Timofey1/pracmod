```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

from scipy.stats import gaussian_kde
```

## Часть 1 

через x и y

```python
r = 10 
n_samples = 1000

X=[]
Y=[]

while len(X) != n_samples:
    x = np.random.uniform(-10, 10)
    y = np.random.uniform(-10, 10)
    if x**2 + y**2 <= r**2:
        X.append(x)
        Y.append(y)
        
```

```python
angels = np.linspace(0,2*np.pi,1000)
cirX = r*np.cos(angels)
cirY = r*np.sin(angels)
```

```python
plt.figure(figsize=(4,4))
plt.plot(cirX,cirY,"r")
plt.plot(X,Y,".k")
```

через угол и расстояние

```python
aX = []
aY = []

while len(aX) != n_samples:
    angl = np.random.uniform(0, 2*np.pi)
    dist = np.random.uniform(0, 10)
    aX.append(dist*np.cos(angl))
    aY.append(dist*np.sin(angl))
```

```python
plt.figure(figsize=(4,4))
plt.plot(cirX,cirY,"r")
plt.plot(aX, aY, ".k")
```

## Часть 2

##### Пункт 1.2

```python
print("Mean x for method 1", np.mean(X))
print("Mean y for method 1", np.mean(Y))
print("Mean x for method 2", np.mean(aX))
print("Mean y for method 2", np.mean(aY))
```

```python
print("Var x for method 1", np.var(X))
print("Var y for method 1", np.var(Y))
print("Var x for method 2", np.var(aX))
print("Var y for method 2", np.var(aY))
```

##### Пункт 1.3

```python
dot_x = 20
dot_y = 20
```

```python
def dist(x1,y1,x2,y2):
    return np.sqrt((x2-x1)**2 + (y2-y1)**2)
```

```python
distances_1 = []
for i in range(n_samples):
    distances_1.append(dist(X[i], Y[i], dot_x, dot_y))
distances_1.sort()

distances_2 = []
for i in range(n_samples):
    distances_2.append(dist(aX[i], aY[i], dot_x, dot_y))
distances_2.sort()
```

```python
# max_x = np.linspace(0, 20, 1000)

# plt.plot(max_x, distances_1)
# plt.plot(max_x, distances_2)
```

```python
kde_dist_1_y = gaussian_kde(distances_1).evaluate(distances_1)
kde_dist_2_y = gaussian_kde(distances_2).evaluate(distances_2)
```

```python
plt.figure(figsize=(8,6))
plt.plot(distances_1, kde_dist_1_y)
plt.hist(distances_1, bins=100, alpha=0.2, density=True)
plt.title("Метод 1")
```

```python
plt.figure(figsize=(8,6))
plt.plot(distances_2, kde_dist_2_y)
plt.hist(distances_2, bins=100, alpha=0.2, density=True)
plt.title("Метод 2")
```

##### Пункт 1.4

```python
distances_1 = []
distances_2 = []

for i in range(n_samples):
    a,b = np.random.randint(0, n_samples-1), np.random.randint(0, n_samples-1)
    distances_1.append(dist(X[a], Y[a], X[b], Y[b]))
    distances_2.append(dist(aX[a], aY[a], aX[b], aY[b]))

distances_1.sort()
distances_2.sort()
```

```python
kde_dist_1_y = gaussian_kde(distances_1).evaluate(distances_1)
kde_dist_2_y = gaussian_kde(distances_2).evaluate(distances_2)
```

```python
plt.figure(figsize=(8,6))
plt.plot(distances_1, kde_dist_1_y)
plt.hist(distances_1, bins=100, alpha=0.2, density=True)
plt.title("Метод 1")
```

```python
plt.figure(figsize=(8,6))
plt.plot(distances_2, kde_dist_2_y)
plt.hist(distances_2, bins=100, alpha=0.2, density=True)
plt.title("Метод 2")
```
